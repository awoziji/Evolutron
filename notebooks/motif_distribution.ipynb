```python
# Special matplotlib import for the Jupyter notebook
%matplotlib inline
from IPython.display import IFrame
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def set_style():
    # This sets reasonable defaults for font size for
    # a figure that will go in a paper
    sns.set_context("paper")
    
    # Set the font to be serif, rather than sans
    sns.set(font='serif')
    
    # Make the background white, and specify the
    # specific font family
    sns.set_style("white", {
        "font.family": "serif",
        "font.serif": ["Times", "Palatino", "serif"]
    })
def set_size(fig):
    fig.set_size_inches(6, 4)
    plt.tight_layout()
```

```python
def plot_motif_dist(lengths):
    
    fig, ax = plt.subplots()
    
    n, bins, patches = ax.hist(lengths, 40, range=(0,200), facecolor='green', alpha=0.75)

    ax.set_xlabel("Motif Length (bins)")
    ax.set_ylabel("Number of motifs")
    
    ax.set_xticks([0,25, 50, 75, 100, 125, 150, 175, 200])
    
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    
    return fig, ax
```

```python
import numpy as np
import os
from Bio import SeqIO
with np.load('lengths.npz') as f:
    lengths = f['lengths']
```

```python
set_style()
fig, axes = plot_motif_dist(lengths)
set_size(fig)
plt.savefig('/Users/thrakar9/Desktop/motif_lens.pdf', dpi=300)
```
